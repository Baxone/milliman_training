```python
import numpy as np

# importe de siniestros anuales por poliza (€)
# np.array me genera un vector, lista de elementos pero numpy
siniestros = np.array([450, 800, 1200, 1500, 2300, 5000])

# exposiciones a riesgo ( años asegurados )
exposicion = np.array([1,1,1,0.5, 2 ,1])

# pesos de cada poliza
pesos = np.array([0.10, 0.15, 0.20, 0.10, 0.25, 0.20])
```

Funciones estadistica de NUMPY:

np.mean - sinestro medio . Coste medio esperado por poliza

```python
np.mean(siniestros)
```

np.median() - siniestro típico - Valor central, menos afectado por siniestros grandes

```python
np.median(siniestros)
```

np.sum - coste total de la cartera - coste total de sinestros

```python
np.sum(siniestros)
```

np.std/np.var - volatilidad del riesgo - medida de incertidumbre

```python
np.std(siniestros)
```

```python
np.var(siniestros)
```

np.percentile/np.quantile - riesgos extremo (VaR) - que casos superan 5% de riesgo extremo

```python
np.percentile(siniestros, 95)
```

```python
np.quantile(siniestros, 0.95)
```

Segmentación de riesgos - Clasificacion automatica de mi riesgo en funcion la prima
np.where

```python
riesgo = np.where(siniestros > 2000, 'Alto', 'Normal')
print(riesgo)
```

np.clip - un vector con limite de cobertura - La asegurado solo paga hasta un maximo de 3000. Como quederia el pago de la prima

```python
pago_limitado = np.clip(siniestros, 0, 3000)
print(pago_limitado)
```

np.maximun - deducción - deducir 500 euros de la poliza

```python
valor_deducible = 500
pago = np.maximum(siniestros - valor_deducible, 0)
print(pago)
```

Media ponderada np.dot(data, ponderacion) - prima con media ponderada

```python
prima_media = np.dot(siniestros, pesos)
print(prima_media)
```

pagos acumulados - Evolución acumulada del coste de los sinistros
np.cumsum(vector)

```python
np.cumsum(siniestros)
```

Frecuencia de sinistros - funcion de poisson - numero de siniestro por poliza al año

```python
np.random.poisson(0.8, 1000)
```

np.power - valor presente a 2 años con un 3% anual

```python
valor_actual = siniestros / np.power(1.03, 2)
print(valor_actual)
```

evolucion de la sinestralidad en el timpo usando como referencias la cantidad de siniestros
np.polifit(años, vector)

```python
anios = np.arange(len(siniestros))
tendencia = np.polyfit(anios, siniestros, 0)
print(tendencia)
```

ranking de riesgos - identificar la polizas mas peligrosas
np.argsort()

```python
ranking = np.argsort(siniestros)[::-1]
print(ranking)
```
